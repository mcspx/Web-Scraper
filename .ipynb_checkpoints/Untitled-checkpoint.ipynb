```python
datas = []
indexs = []
cols = ['count','name']
n = 0
for p in political:
    n = n + 1
    datas.append([p['count'],p['name']])
    indexs.append(p['name'])
    #print(p)
```

```python
df = pd.DataFrame(datas,index=indexs,columns=cols)
```

```python
df
```

```python
barchart = df.plot(kind='bar',title='Total count word of Political on website.',legend=None,figsize=(15, 15))
barchart.set_ylabel('Count word')
barchart.set_xlabel('Political')
```
