```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.display import HTML

jdata = pd.read_json('jdata.json')
%matplotlib inline
political = jdata.political_data
datas = []
indexs = []
cols = ['name','count']
n = 0
for p in political:
    n = n + 1
    datas.append([p['name'],p['count']])
    indexs.append(n)
    #print(p)
df = pd.DataFrame(datas,index=indexs,columns=cols)
barchart = df.plot(kind='bar',title='Total count word of Political on website.',legend=None,figsize=(20, 15))
barchart.set_ylabel('Count word')
barchart.set_xlabel('Political')
display(HTML(df.to_html()))
```
