```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

jdata = pd.read_json('jdata.json')
%matplotlib inline
political = jdata.political_data
datas = []
indexs = []
cols = ['count','name']
n = 0
for p in political:
    n = n + 1
    datas.append([p['count'],p['name']])
    indexs.append(p['name'])
    #print(p)
df = pd.DataFrame(datas,index=indexs,columns=cols)
barchart = df.plot(kind='bar',title='Total count word of Political on website.',legend=None,figsize=(13, 13))
barchart.set_ylabel('Count word')
barchart.set_xlabel('Political')
```
